```python
import matplotlib.pyplot as plt
df.columns = df.columns.str.strip() 
top_10 = df.sort_values(by='Market Cap(in B USD)', ascending=False).head(10) 
plt.figure(figsize=(10, 6))
plt.bar(top_10['Company'], top_10['Market Cap(in B USD)'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Market Cap (in Billions USD)')
plt.title('Top 10 Companies by Market Cap (2023)')
plt.tight_layout()
plt.show()
```
